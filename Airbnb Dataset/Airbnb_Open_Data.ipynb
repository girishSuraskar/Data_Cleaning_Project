# Renaming the columns

```python
data = pd.read_csv('Airbnb_Open_Data.csv')
```

```python
data.columns
```

```python
data.rename(columns = str.upper,inplace=True )  #converting column name to upper case
print(data.columns)
# now replacing space " " with underscore "_" using map
data.columns =  data.columns.map(lambda col : col.replace(" ","_"))
print(data.columns)
```

# Dropping Duplicates

```python
# Dulicated or not values counting
print(data.duplicated().value_counts())

# Total duplicated rows
print(data.duplicated().sum())

data.shape
```

```python
# Dropping duplicated rows
data.drop_duplicates(inplace =True)
print(data.duplicated().sum()) #verifying
data.shape
```

# Removing the NA values from the dataset

```python
data.isna().sum() #column wise na values count
```

```python
# we are droping LAST_REVIEW column
data.drop(columns = ['LAST_REVIEW'],inplace=True)
data.shape
```

# Dropping NA 
As our data has so many rows which contain at least one na values so we can't directly use dropna()

```python
data = data.apply(data.select_dtypes(include=['object']).fillna("No record"))
```

```python

# Filling NA with mean by handling non number columns
data = data.fillna(data.select_dtypes(include=['number']).mean())
data
```

# Removing '$' sign from price field

```python
print("Before cleaning 'PRICE' columns: ")
data['PRICE']
```

```python
print("After cleaning 'PRICE' columns: ")
data['PRICE'] = data['PRICE'].str.replace("$","")
data['PRICE']
```

```python
print("Now we need to remove comma ")
data['PRICE'] = data['PRICE'].astype(str)
data['PRICE'] = data['PRICE'].str.replace(",", "", regex=True)
print(data['PRICE']) #verifying  
```

```python

print("Type of data: ",type(data['PRICE'][1]))
print("We need to convert it to numeric type")
data['PRICE']=pd.to_numeric(data['PRICE'],errors='coerce')
data['PRICE'] = data['PRICE'].fillna(0).astype(int)
type(data['PRICE'][1] )
```

```python

data.to_csv("Airbnb_Cleaned_DATA",index = False)
```
