```python
columns_to_keep = [ 'NAME', 'host id', 'host_identity_verified', 'host name',
       'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country',
       'country code', 'instant_bookable', 'cancellation_policy', 'room type',
       'Construction year', 'price', 'service fee', 'minimum nights',
       'number of reviews', 'last review']
columns_to_drop = ['id','reviews per month',
       'review rate number', 'calculated host listings count',
       'availability 365', 'house_rules', 'license']
print(f"Number of columns to keep : {len(columns_to_keep)}")
print(f"Number of columns to drop : {len(columns_to_drop)}")
```

```python
# 1st method Filtering data
new_df = df[columns_to_keep]
new_df
```

```python
# 2nd Method, Dropping Columns
df.drop(columns = columns_to_drop,inplace = True)
df
```

# Renaming the columns

```python
data = pd.read_csv('Airbnb_Open_Data.csv')
```

```python
data.columns
```

```python
data.rename(columns = str.upper,inplace=True )  #converting column name to upper case
print(data.columns)
# now replacing space " " with underscore "_" using map
data.columns =  data.columns.map(lambda col : col.replace(" ","_"))
print(data.columns)
```

# Dropping Duplicates

```python
# Dulicated or not values counting
print(data.duplicated().value_counts())

# Total duplicated rows
print(data.duplicated().sum())

```

```python
# Dropping duplicated rows
data.drop_duplicates(inplace =True)
print(data.duplicated().sum()) #verifying
```

# Removing the NA values from the dataset

```python
data.isna().sum() #column wise na values count
```

```python
# we are droping LAST_REVIEW column
data.drop(columns = ['LAST_REVIEW'],inplace=True)
```

```python
# Dropping NA
data_without_na = data.dropna()
```

```python

# Filling NA with mean by handling non number columns
data_replce_na_with_mean = data.fillna(data.select_dtypes(include=['number']).mean())
data_replce_na_with_mean
```

```python
# In 'HOST_IDENTITY_VERIFIED' column we need to change unconfirmed	to -1 ,nan to 0 verifed to 1
print(data['INSTANT_BOOKABLE'])
data_replce_na_with_mean['INSTANT_BOOKABLE']= data_replce_na_with_mean['INSTANT_BOOKABLE'].apply(lambda x: 1 if x== True else 0)
print(data_replce_na_with_mean['INSTANT_BOOKABLE'])
```
